Find Duplicates Example:

```
    students = [('jack', 34, 'Sydeny'),
                ('Riti', 30, 'Delhi'),
                ('Aadi', 16, 'New York'),
                ('Riti', 30, 'Delhi'),
                ('Riti', 30, 'Delhi'),
                ('Riti', 30, 'Mumbai'),
                ('Aadi', 40, 'London'),
                ('Sachin', 30, 'Delhi')
                ]                
dfObj = pd.DataFrame(students, columns=['Name', 'Age', 'City'])
print(dfObj)
# Select duplicate rows except first occurrence based on all columns
print("All duplicate rows except first occurrence based on all columns")
duplicateRowsDF = dfObj[dfObj.duplicated()]
print(duplicateRowsDF)

# Select all duplicate rows based on all columns
duplicateRowsDF = dfObj[dfObj.duplicated(keep=False)]
print("All Duplicate Rows based on all columns are :")
print(duplicateRowsDF)

# Select all duplicate rows based on multiple column names in list
duplicateRowsDF = dfObj[dfObj.duplicated(['Name','Age', 'City'], keep=False)]
print("All Duplicate Rows based on multiple column names in list are:")
print(duplicateRowsDF)
```

```
data_explore = df[['fechacierre.x',"CUNTID.x","Rating","Score2","Score","Delinq_31 a 60"]]
num_clients = data_explore.groupby("fechacierre.x").agg({"CUNTID.x": pd.Series.nunique}) #contabiliza los id √∫niocs por a√±o
num_clients.rename(columns = {"CUNTID.x": "num_unique_id"}, inplace=True)
num_clients.head()
```
